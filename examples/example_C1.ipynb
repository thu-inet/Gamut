This example gives the demonstration of some marginal capacities of gamspec.

```python
import numpy as np
import matplotlib.pyplot as plt

from gamspec import Spectrum, Pipe
```

`gamspec.operators.BasicOperator.Functionor` is a simple wrapper for scalar functions of single input.

```python
from gamspec.operators.BasicOperator import Functionor

spec = Spectrum(np.linspace(0, 10, 100))

modulo = Functionor(func=lambda x: x % 2, label='modulo')
divide = Functionor(func=lambda x: x / 2, label='divide')

modulo_output =  modulo(spec)
divide_output =  divide(spec)

modulo_output.plot()
divide_output.plot()
plt.ylim(0, 5)
plt.legend()
```

`gamspec.operators.BasicOperator.Combinor` is a simple wrapper for scalar functions of multiple inputs.

```python
from gamspec.operators.BasicOperator import Combinor, Multiplier

spec1 = Spectrum(np.linspace(0, 4, 100))
spec2 = Spectrum(np.linspace(0, 4, 100))

multiply = Multiplier()
exponent = Combinor(func=lambda x, y: x ** y, label='exponent', inp_num=2)

multiply_output = multiply([spec1, spec2])
exponent_output = exponent([spec1, spec2])

multiply_output.plot()
exponent_output.plot()
plt.ylim(0, 100)
plt.legend()
```

`gamspec.operators.BasicOperator.Convolutionor` is a wrapper of `numpy.conv`, so it is usually used to construct linear filters or smoothers.

```python
from gamspec.operators.BasicOperator import Convolutionor
from gamspec import SimulatedSpectrum, simuspecs


spec = SimulatedSpectrum(**simuspecs['simple'])

conv1 = Convolutionor(kernel=[1/4, 1/2, 1/4], label='conv1')
conv2 = Convolutionor(kernel=[1/8, 1/8, 1/2, 1/8, 1/8], label='conv2')

conv1_output = conv1(conv1(spec)) - 30
conv2_output = conv2(conv2(spec)) + 30

spec.plot()
conv1_output.plot()
conv2_output.plot()
plt.legend()
```
